## Pre-processing Sales 

```python
df_sales
```

```python
df_sales.columns
```

```python
df_sales.isnull().sum()
```

```python
df_sales.drop(columns=['Address Number','Backlog Amount',
       'BackOrder Amount','DateKey','Family Class','Doc Type',
                       'Item Class', 'Item Number','Line Number','Line Type',
                       'Parent Address Number','Unit Price','Payment Terms','Promised Delivery Date'],inplace=True, axis=1)
```

```python
df_sales.drop(columns=['Actual Delivery Date','Invoice Number','Line Desc 1','Order Number','U/M'],inplace=True, axis=1)
```

```python
df_sales = pd.merge(df_sales, df_srep, how='left', left_on='Sales Rep', right_on='Sales Rep')
```

```python
df_sales
```

```python
df_sales.drop_duplicates(inplace=True)
```

```python
df_sales.isnull().sum()
```

```python
df_sclean = df_sales[df_sales['Sales Amount'].notna()]
```

```python
df_sclean.isnull().sum()
```

```python
df_sclean["Discount Amount"] = df_sclean["Discount Amount"].fillna(0)
```

```python
df_sclean.isnull().sum()
```

```python
df_sclean = df_sclean[df_sclean['Sales Price'].notna()]
df_sclean.isnull().sum()
```

```python
df_sclean.shape
```

```python
df_sclean.columns
```

```python
df_sclean = df_sclean[['CustKey', 'Item-Branch Key', 'Invoice Date',
       'List Price','Sales Quantity', 'Sales Amount', 'Sales Amount Based on List Price',
       'Sales Cost Amount', 'Sales Margin Amount', 'Sales Price',
       'Sales Unit Cost Price','Discount Amount',
       'Sales Rep Name','Sales Rep']]
df_sclean.head()
```

```python
df_sclean["Sales Rep"]=df_sclean["Sales Rep Name"]+" "+df_sclean["Sales Rep"].map(str)
```

```python
df_sclean.drop(columns=['Sales Rep Name'],axis=1, inplace=True)
```
