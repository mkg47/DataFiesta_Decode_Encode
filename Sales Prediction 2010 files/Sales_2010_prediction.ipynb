```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('DateSalesMargin.xlsx')
```

```python
df
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
from xgboost import XGBRegressor
model1=XGBRegressor()
model1.fit(X,y)
```

```python
from sklearn.ensemble import RandomForestRegressor
model2 = RandomForestRegressor(n_estimators = 10, random_state = 0)
model2.fit(X, y)
```

```python
def predictx( X_new ):
    p1=model1.predict(X_new)
    p2=model2.predict(X_new)
    p3=[]
    for x in range(len(p1)):
        p3.append((p1[x]+p2[x])/2)
    return p3
```

```python
X_2010=[[2010, 1],
        [2010, 2],
        [2010, 3],
        [2010, 4],
        [2010, 5],
        [2010, 6],
        [2010, 7],
        [2010, 8],
        [2010, 9],
        [2010, 10],
        [2010, 11],
        [2010, 12]]
X_2010=np.array(X_2010)
X_2010
```

```python
y_2010=predictx(X_2010)
```

```python
Sales2010=pd.DataFrame(data=X_2010,columns=["Year","Month"])
Sales2010["Sales_Margin"]=y_2010
```
